```python

print(data.isnull().sum())


data = data.dropna(subset=['capacity_mw', 'primary_fuel'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

 
sns.countplot(data=data, x='primary_fuel')
plt.title('Distribution of Power Plants by Primary Fuel')
plt.xticks(rotation=45)
plt.show()


sns.histplot(data['capacity_mw'], kde=True)
plt.title('Distribution of Power Plant Capacities')
plt.xlabel('Capacity (MW)')
plt.show()
```

```python
country_capacity = data.groupby('country')['capacity_mw'].sum().sort_values(ascending=False)
country_capacity.head(10).plot(kind='bar', color='green')
plt.title('Top Countries by Total Power Plant Capacity')
plt.xlabel('Country')
plt.ylabel('Total Capacity (MW)')
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
geo_df = gpd.GeoDataFrame(data, geometry=geometry)


geo_df.plot(marker='o', color='red', markersize=5)
plt.show()
```
